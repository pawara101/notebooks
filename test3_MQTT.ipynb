## Preprocessing

```python
# Get list of categorical variables
s = (test30_aug.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)

object_cols.pop() ## Drop Target column

print(object_cols)
```

```python
# Select numerical columns
numerical_cols = [cname for cname in test30_aug.columns if test30_aug[cname].dtype in ['int64', 'float64']]

print(numerical_cols)
```

```python
## Unique value
for cols in numerical_cols:
  print(f"{cols} :",test30_aug[cols].unique())
```

```python
low_cardinality_cols = [col for col in object_cols if test30_aug[col].nunique() < 15]
```

```python
low_cardinality_cols
```

```python
## Encoding target for easiness
test30_aug.loc[test30_aug['target']=='dos','target']=1.0
test30_aug.loc[test30_aug['target']=='slowite','target']=2.0
test30_aug.loc[test30_aug['target']=='flood','target']=3.0
test30_aug.loc[test30_aug['target']=='bruteforce','target']=4.0
test30_aug.loc[test30_aug['target']=='malformed','target']=5.0
test30_aug.loc[test30_aug['target']=='legitimate','target']=6.0
```

```python
## Object/categorical columns

## Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder

# Make copy to avoid changing original data 
label_test30_aug = test30_aug.copy()

# Apply ordinal encoder to each column with categorical data
ordinal_encoder = OrdinalEncoder()

label_test30_aug[object_cols] = ordinal_encoder.fit_transform(label_test30_aug[object_cols])
```

## LGBM Classifier

```python
import lightgbm as lgb
import time
clf = lgb.LGBMClassifier(learning_rate= 0.08, 
                            max_depth= 5, 
                            n_estimators=50, 
                            num_leaves= 31,
                            boosting_type= 'goss', 
                            colsample_bytree= 1.0,
                            reg_lambda= 0.5)
start = time.time()
clf.fit(X_train[GA_selected_feat_25],y_train)
stop = time.time()

print(f"Training time: {stop - start}s")

y_pred=clf.predict(X_test[GA_selected_feat_25])

# view accuracy
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred, y_test)
print('LightGBM Model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, y_pred)))

## Confusion Matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))

from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
plot_confusion_matrix(clf, X_test[GA_selected_feat_25], y_test)  
plt.show()
```
